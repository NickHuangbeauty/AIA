```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family']='SimHei' #顯示中文

%matplotlib inline
```

```python
import os
input_path="/data/examples/trend/data/query_log/"
files = []
for _root,_dirs,_files in os.walk(input_path):
    files=_files
files.sort() 
files
```

```python
columnNames = ['FileID','CustomerID','QueryTs','ProductID']
all_data = pd.DataFrame(columns=columnNames)
i=0
for fn in files:
    tmp = pd.read_csv(input_path + fn)
    tmp.columns = columnNames
    all_data = all_data.append(tmp,ignore_index=True)
#    i+=1
#    if i==5:
       break
all_data['times']=1
```

```python
import datetime
q_gwd=all_data.groupby(['FileID'],as_index=False)[['times']].sum()
q_gwd=q_gwd.where(q_gwd > 50).dropna(axis=0, how='any')
q_gwd2 = q_gwd.drop('times', axis = 1)
q_gwd2['open_50times']= 1
q_gwd2
result = pd.merge(all_data,q_gwd2,on='FileID',how='left')

result['open_50times'].fillna(0, inplace=True)
result[['open_50times']] = result[['open_50times']].astype(int)
result.sort()
result
```

```python
train_label = pd.read_csv('/data/examples/trend/data/training-set.csv', encoding = "utf-8", header=None)
test_label = pd.read_csv('/data/examples/trend/data/testing-set.csv', encoding = "utf-8", header=None)
train_label.columns=['FileID','label']
train_label
```

```python
result1 = result.drop('ProductID', axis = 1)
result1 = result1.drop('CustimerID', axis = 1)
result1 = result1.drop('times', axis = 1)
result1 = result1.drop('QueryTs', axis = 1)
result2=result1.drop_duplicates('FileID')
result2
result3 = pd.merge(result2,train_label,on='FileID',how='inner')
```

```python
#result3.to_csv('result3.csv', index=False)
test_label.columns=['FileID','label']
test_label
test_label = test_label.drop('label', axis = 1)
test_label
#result4 = pd.merge(result3,test_label,on='FileID',how='inner')
#result4
```

```python
train_label2= train_label.drop('FileID', axis = 1)
result4= result3.drop('FileID', axis = 1)
result4= result4.drop('label', axis = 1)
```

```python
from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
% pylab inline
#result4 = pd.merge(result3,test_label,on='FileID',how='left')
X_train, X_test, Y_train, Y_test = train_test_split(result4.as_matrix(), train_label2.as_matrix(), test_size=0.3)
```

```python
Y_train.shape
```

```python
eval_set =  [(X_test, Y_test)]

print('start modelling!')
model = XGBClassifier()
model.fit(X_train, Y_train, early_stopping_rounds=10, eval_metric="auc", eval_set=eval_set,verbose=True)
```

```python
#X_pred = model.predict(test_label.as_matrix())
result5 = pd.merge(result2,test_label,on='FileID',how='inner')
result6 = result5.drop('FileID', axis = 1)
result6
X_pred = model.predict_proba(result6.as_matrix())
```
